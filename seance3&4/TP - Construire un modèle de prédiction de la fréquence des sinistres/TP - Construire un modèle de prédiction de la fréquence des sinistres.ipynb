# TP - Construire un modèle de prédiction de la fréquence des sinistres

Le but du projet est de construire un modèle prédictif de la fréquence de survenance des sinistres pour un portefeuille d'assurances auto. Les données étudiées sont téléchargeable à l'adresse : http://freakonometrics.hypotheses.org/50078

Pour un portefeuille d'assurances auto comprenant $n$ polices, soient $(S_i)_{i=1,..,n}$ les variables aléatoires qui représentent le coût de la sinistralité annuelle totale de la police $i$ (pour une année donnée), $M_i$ le nombre de sinistres survenus pour la police $i$ pendant l'année, et $X_i$ le vecteur de caractéristiques du contrat.

Alors on peut écrire que :

$$S_i = \sum_{j=1}^{M_i} S_{i,j}$$

avec $(S_{i,j})_{\substack{i = 1,..,n \\ j = 1,..,M_i}}$ les coûts individuels des sinistres.

## Hypothèses
1) En assurance auto, on peut supposer qu'il n'y a pas de lien entre les sinistres survenus pour des polices distinctes. Ainsi, on fait l'hypothèse que les groupes de variables $(S_i, M_i, X_i, (S_{i,j})_{j = 1,..,M_i})_{i=1,..,n}$ sont i.i.d. pour des $i$ distincts.

2) On fait également l'hypothèse suivante pour tout $i$ fixé : $\forall j = 1,..,M_i, ~~ S_{i,j} ~\unicode{x2AEB}~ M_i ~|~ X_i$. C'est à dire que si l'on connaissait $X_i$, alors connaitre en plus $M_i$ ne nous donnerait pas plus d'information sur les $S_{i,j}$.

3) On suppose enfin que pour tout $i$ fixé, les variable $(S_{i,j})_{j = 1,..,M_i}$ sont i.i.d. C'est à dire que l'on suppose que les coûts des sinistres individuels sont indépendants et de même distribution pour une police données.

## Modèle fréquence * coût moyen

Sous les hypothèses du paragraphe précédent, on peut faire le calcul suivant :

\begin{align*}
E[S_i |X_i] & = E\left[ \sum_{j=1}^{M_i} S_{i,j}~ | ~X_i \right] \\
& = E \left[ E \left[ \sum_{j=1}^{M_i} S_{i,j}~ | ~M_i, X_i \right] |~X_i\right] \\
& = E \left[ \sum_{j=1}^{M_i} E \left[ S_{i,j}~ | ~M_i, X_i \right] |~X_i\right] \\
& = E \left[ \sum_{j=1}^{M_i} E \left[ S_{i,j}~ | ~X_i \right] |~X_i\right] (hyp. 2))\\
& = E \left[ M_i \cdot E \left[ S_{i,j}~ | ~X_i \right] |~X_i\right] (hyp. 3))\\
& = \underset{\text{fréquence}}{\underbrace{E \left[ M_i ~|~X_i\right]}} \cdot \underset{\text{coût moyen}}{\underbrace{E \left[ S_{i,j}~ | ~X_i \right]}}
\end{align*}

Ce qui explique la modélisation sous la forme Fréquence * coût moyen
