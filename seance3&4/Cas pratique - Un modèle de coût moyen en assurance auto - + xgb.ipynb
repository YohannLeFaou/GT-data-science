# Construction de la base d'apprentissage

Nous avons chargé les données et, à travers quelques statistiques descriptives, nous avons vérifié que les données sont exploitables. 

Notre but est de construire un modèle statistique qui servirait à prédire le coût d'un sinistre. Pour cela, nous disposons :

* d'une base sinistre (`data_claim0`) : cette base regroupe les informations concernant 14243 sinistres. Chaque sinistre est associé à un triplet\* : client - véhicule - identifiant-de-sinistre (l'identifiant de sinistre permet de distinguer 2 sinistres assossiés à un même couple client-véhicule). C'est à partir des observations de montants de sinistres que nous connaissons que nous allons calibrer notre modèle de coût : **la variable *claim amount* sera la *variable à expliquer* de notre modèle**.
* d'une base prospect (`data_propect0`) : cette base contient des informations concernant les 100000 clients du portefeuille, y compris les clients qui n'ont pas eu d'accident. C'est en faisant correspondre les informations clients et les montants de sinistres associés que nous pourrons identifier des facteurs impactant le coût moyen de sinisitre : **les variables de la base *data_prospect0* seront les variables explicatives de notre modèle**.

<br>
\*inutile de tenir compte de la variable `id_year` qui vaut toujours `0` l'année zéro.

<br>
Pour calibrer un modèle statistique, il est nécessaire de mettre en regard des variables explicatives, avec une (ou parfois plusieurs) variables à expliquer.

Nous devons donc faire correspondre les informations contenues dans les 2 bases. Plus précisement, pour chaque sinistre de notre base *sinitres*, nous  allons faire correspondre les informations relatives au couple *client-véhicule* concerné.

En informatique, cela correspond à faire un *merge*
de bases de données.

## Merge de la base sinistre et de la base prospect

*pandas* fournit tous les outils nécessaire pour réaliser un merge de bases de données.

Pour faire correspondre 2 bases de données, il faut également disposer d'une *clé* (ou *identifiant*) qui permet de faire la jointure. Dans notre cas la clé est le couple *client-véhicule*.

On supprime d'abord la colonne *id_year* de la base *data_claim0* :

```python
data_claim0 = data_claim0.drop("id_year", axis=1)
```

On vérifie qu'il n'y a pas de doublon dans la base *data_prospect0* :

```python
data_prospect0[["id_client", "id_vehicle"]].drop_duplicates().shape # ok, pas de doublons
```

On fait le merge :

```python
base_cout = data_claim0.merge(right=data_prospect0,
                               how="inner",
                               on=["id_client","id_vehicle"])
```

On vérifie que le merge a bien marché :

```python
base_cout.shape
```

```python
base_cout.head()
```

Tout semble être en ordre
