# Modélisation statistique

## Cadre général

### Régression et classement

On note $Y$ la variable `claim_amount` : $Y$ est la variable que l'on souhaite modéliser. $Y$ est une variable quantitative donc construire un modèle pour expliquer $Y$ s'appelle faire une **régression**.

Si $Y$ était une variable qualitative, on dirait que prédire $Y$ est un problème de **classement** (classement **binaire** : 2 classes, ou classement **multiple** : 3 classes ou plus).

On note $X$ le vecteur des variables explicatives utilisées dans un modèle. 


### Différents type d'apprentissage statistique

Ici, le modèle vise à expliquer une variable $Y$ bien particulière, qui est le coût d'un sinistre. On parle alors d'**apprentissage supervisé**. On dit que $Y$ est la variable cible (ou variable à expliquer) 

Il arrive qu'aucune variable ne joue un rôle particulier : c'est le cas par exemple pour un problème de classification, ou le but d'un modèle est de segmenter un échantillon en classes homogènes. On parle alors de problèmes d'**apprentissage non supervisé**.

Plus généralement, le but de l'apprentissage non supervisé est de créer de la connaissance à partir de données, sans qu'aucune variable ne joue le rôle de variable cible (par exemple on dit que l'humain apprend de manière non supervisée)

Le 3e type d'apprentissage statistique, de plus en plus courant, est l'**apprentissage par renforcement** (exemples : voiture autonome, robots qui jouent à des jeux vidéo)

### Principe de la régression

Modèle du type : $Y = f(X) + \epsilon$

* Avec
  - $Y$ : variable à expliquer (souvent 1 seule dimension mais ce n’est pas obligatoire)
  - $X$ :  vecteur de variables explicatives $(X \in R^d)$
  - $\epsilon$ : bruit (selon le modèle, variable aléatoire centrée/réduite/gaussienne)

Pour calibrer les modèles on se base sur des observations $(y_i, x_i)_{i = 1,..,n}$ où les réalisations $i$ sont supposées indépendantes

On note

* $\hat{y}_i$ la prédiction du modèle pour l’observation $i$
* $\epsilon_i = y_i - \hat{y}_i$ : le résidu du modèle pour l’individu $i$

## GLM (Generalised Linear Model)

### Explications sur le GLM

Commençons par l'exemple du modèle linéaire gaussien.

Le modèle linéaire (LM), ou modèle linéaire gaussien, est un cas particulier du GLM.

**Modèle linéaire (gaussien)** :

* $Y = ~ ^t \beta \cdot X + \epsilon$ avec :
  - $\beta = (\beta_1, ..., \beta_d)$ un vecteur de paramètres
  - $\epsilon \sim N(0, \sigma^2)$
  
En particulier, $E[Y|X] = ~ ^t \beta \cdot X $. De plus on suppose que la loi de $Y$ sachant $X$ est une gaussienne : $\mathcal{L}(Y|X) \sim N(^t\beta \cdot X, \sigma^2)$

**Modèle linéaire généralisé (GLM)** :

* $g(E[Y|X]) = ~ ^t \beta \cdot X$
* Les lois $\mathcal{L}(Y|X)$ appartiennent toutes à la même distribution de la famille exponentielle (Bernouilli, Binomiale, Poisson, Binomiale négative, Uniforme, Gaussienne, Gamma, Exponentielle, et d'autres encore...)

avec :

* $\beta = (\beta_1, ..., \beta_d)$ un vecteur de paramètres
* $g$ une **fonction lien**

Remarque : 
 - LM est bien un cas particulier du GLM, avec $g = identité$ et les loi $\mathcal{L}(Y|X)$ qui sont gaussiennes

### Autres exemples de GLM classiques

**GLM Poisson** (avec lien log) :

* $\mathcal{L}(Y|X) \sim Poisson(e^{^t \beta \cdot X})$

utilisé avec $Y$ une variable de comptage. Exemple : modélisation de la fréquence des sinistres.

**GLM  Gamma** (avec lien log) :

* $\mathcal{L}(Y|X) \sim Gamma(e^{^t \beta \cdot X}, 1)$

utilisé avec $Y$ une variable à valeurs positives. Exemple : modélisation d'un coût

**GLM binomiale** (avec lien logit : $g(u) = log\left( \frac{u}{1-u} \right)$)

* $\mathcal{L}(Y|X) \sim Bernouilli( \frac{e^{^t \beta \cdot X}}{1 + e^{^t \beta \cdot X}})$

Remarque :
 - Pour une distribution donnée, il existe un nombre limité de fonctions lien possibles

### Calibrage d'un GLM

Calibrer un GLM consiste à estimer le vecteur de paramètres $\beta$.

Pour cela, la technique est de trouver le $\beta$ qui maximise la vraisemblance de l'échantillon

Exemple : dans le cas du modèle linéaire, cela revient à trouver $\beta^*$ qui vérifie :

$$ \beta^* = \underset{\beta \in R^d}{\text{argmin}} \sum_{i = 0}^n (y_i - ^t \beta \cdot x_i )^2 $$

### Evaluation d'un modèle par validation croisée

Il existe des stratégies qui permettent d'estimer les performances des modèles comme ci ces derniers étaient en phase de test. Ces différentes stratégies sont basées sur le ré-échantillonage des données. La validation croisée est l'une de ces méthodes.

```python
print(1 + cv_scores_tree2.mean() / base_cout3.claim_amount.std()**2) # R2 test
```

Nous utilisons ici la classe `RandomForestRegressor` qui implémente la régression par forêt aléatoire.

Remarque : en Python on la struture classqiue d'implémentation des algorithmes est : un algorithme = une classe. La classe comporte alors 3 éléments principaux

* un constructeur, pour créer les objets de la classe (il porte le même nom que la classe)
* une méthode `fit`, pour calibrer le modèle
* une méthode `predict`, pour utiliser le modèle sur de nouvelles données

### Critères de précision des modèles :

#### Score train

R2 train :

```python
rf1.score(X=base_cout3[x_vars_forest], y=base_cout3.claim_amount) # R2 train
```

#### Score Out Of Bag (OOB)

Lorsqu'une observation $(y_i, x_i)$ ne fait pas partie d'un échantillon bootstrap $D_j$ (out of bag), alors la prédiction OOB de l'arbre $T_j$ est dite prédiction **out of bag**. Cette prédiction correspond à une prédiction "test" car l'observation $i$ n'a pas été utilisée pour construire l'arbre $T_j$.

La prédiction OOB d'une forêt pour une observation $i$ correspond à la moyenne de ses prédictions OOB parmi les arbre $(T_j){j=1,..,M}$

Remarque : il y a une certaine probabilité qu'une observation $i$ appaarisse dans tous les échantillons bootstrap, et n'est pas de prédiction OOB. Mais en pratique pour $m \ge 100$, cette probabilité est très faible.

```python
print(rf1.oob_prediction_) # predictions OOB
```

L'erreur out of bag d'une forêt aléatoire est assimilable à une erreur test.

Elle est pratique car elle peut être calculée directement lors de la construction de l'arbre (paramètre `oob_score` de la fonction `RandomForestRegressor`).

Toutefois, l'erreur OOB est moins fiable que l'erreur de validation croisée.

```python
print(rf1.oob_score_) # R2 out of bag
```

#### Score test : Cross validation

On peut là encore utiliser la fonction `cross_val_score` de `sklearn`pour rapidement évaluer par **cross validation** la performance test d'un modèle.

Dans notre exemple, le score out of bag de la forêt aléatoire est proche du score obtenu par validation croisée

```python
# R2 par cross validation
cv_scores_rf1 = cross_val_score(estimator=rf1,
                                y=base_cout3.claim_amount,
                                X=base_cout3[x_vars_forest],
                                scoring = "neg_mean_squared_error",
                                cv=10)
print(cv_scores_rf1) # "neg_mean_squared_error" pris sur chacun des 10 échantillons
```

```python
print(1 + cv_scores_rf1.mean() / base_cout3.claim_amount.std()**2) # R2 test
```

### Importance des variables

L'un des output intéressant de la forêt aléatoire concerne l'estimation de l'importance des variables explicatives dans le modèle. Dans la classe `RandomForestRegressor`, l'importance des variables est estimée par le principe du "`mean decrease impurity`".

La perte d'impureté en un noeud d'un arbre est égale à 

$$|f| \cdot Var(f) - \left( |f_g| \cdot Var(f_g) + |f_d| \cdot Var(f_d) \right)$$

Pour une variable $j \in [[ 1, d ]]$ donnée et un arbre de régression $T$, l'importance de la variable $j$ dans l'arbre $T$ correspond à la somme des pertes d'impureté des noeuds où la coupure est faite selon la variable $j$. Par suite, l'importance de la variable $j$ dans une forêt aléatoire est la moyenne des importances de la variable $j$ dans chaque arbre.

L'importance des variables de la classe `RandomForestRegressor` correspond donc aux résultats de ce calcul, au détail près que les importances des variables sont normées de sorte qu'elles somment à 1.

Remarque : On peut aussi définir par cette méthode l'importance des variables pour un arbre CART unique, mais cette mesure est peu fiable car calculée à partir d'un seul arbre.

```python
importance_rf1 = pd.DataFrame({"importance" : rf1.feature_importances_,
                               "variable" : x_vars_forest})
# appeler "rf1.feature_importances_" pour obtenir l'importance des variables pour la forêt aléatoire
importance_rf1.sort_values("importance", ascending=False).head(10)
```
