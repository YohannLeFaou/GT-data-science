# Construction de la base d'apprentissage

Nous avons chargé les données et, à travers quelques statistiques descriptives, nous avons vérifié que les données sont exploitables. 

Notre but est de construire un modèle statistique qui servirait à **prédire le coût d'un sinistre**. Pour cela, nous disposons :

* d'une base sinistre (`data_claim0`) : cette base regroupe les informations concernant 14243 sinistres. Chaque sinistre est associé à un triplet\* : client - véhicule - identifiant-de-sinistre (l'identifiant de sinistre permet de distinguer 2 sinistres assossiés à un même couple client-véhicule). C'est à partir des observations des montants de sinistres que nous connaissons que nous allons calibrer notre modèle de coût : **la variable *claim amount* sera la *variable à expliquer* de notre modèle**.
* d'une base prospect (`data_propect0`) : cette base contient des informations concernant les 100000 clients du portefeuille, y compris les clients qui n'ont pas eu d'accident. C'est en faisant correspondre les informations clients et les montants de sinistres associés que nous pourrons identifier des facteurs impactant le coût moyen de sinisitre : **les variables de la base *data_prospect0* seront les variables explicatives de notre modèle**.

<br>
\*inutile de tenir compte de la variable `id_year` qui vaut toujours `0` l'année zéro.

<br>
Pour calibrer un modèle statistique, il est nécessaire de mettre en regard des variables explicatives, avec une (ou parfois plusieurs) variables à expliquer.

Nous devons donc faire correspondre les informations contenues dans les 2 bases. Plus précisement, pour chaque sinistre de notre base *sinitres*, nous  allons faire correspondre les informations relatives au couple *client-véhicule* concerné.

En informatique, cela correspond à faire un *merge*
de bases de données.

## Merge de la base sinistre et de la base prospect

*pandas* fournit tous les outils nécessaire pour réaliser un merge de bases de données.

Pour faire correspondre 2 bases de données, il faut également disposer d'une *clé* (ou *identifiant*) qui permet de faire la jointure. Dans notre cas la clé est le couple *client-véhicule*.

On supprime d'abord la colonne *id_year* de la base *data_claim0* :

```python
data_claim0 = data_claim0.drop("id_year", axis=1)
```

On vérifie qu'il n'y a pas de doublon dans la base *data_prospect0* :

```python
data_prospect0[["id_client", "id_vehicle"]].drop_duplicates().shape # ok, pas de doublons
```

On fait le merge :

```python
base_cout = data_claim0.merge(right=data_prospect0,
                               how="inner",
                               on=["id_client","id_vehicle"])
```

On vérifie que le merge a bien marché :

```python
base_cout.shape
```

```python
base_cout.head()
```

Tout semble être en ordre

### Création de fonctions pour la visualisationde données

```python
def regression_graph_quali_var(data, y_var, x_var, ylim = None) :

    # Construction de l'objet de la classe figure qui représente l'image entière
    # et des 2 objets de classe axes, qui correspondent aux 2 graphiques
    fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(6,8)) 

    # premier graphique, objet ax0
    data.boxplot(column=[y_var],
                      by=x_var, 
                      showmeans=True, 
                      showfliers=False, 
                      meanline=True,
                     ax = ax0) 
    ax0.title.set_visible(False)
    ax0.xaxis.label.set_visible(False)
    if ylim is not None:
        ax0.set_ylim(ylim)
        
    # second graphique : objet ax1
    data.groupby(x_var).size().plot(kind="bar", ax=ax1)
    ax1.set_xlabel(xlabel=x_var, fontsize=12)
    ax1.grid(axis="y")
    ax1.set_axisbelow(True)

    # modification du titre figure : objet 'figure'
    fig.suptitle("Variable {}".format(x_var), fontsize=18)

```

```python
def regression_graph_quanti_var(data, y_var, x_var, type_x, ylim = None, xlim = None) :

    # Construction de l'objet de la classe figure qui représente l'image entière
    # et des 2 objets de classe axes, qui correspondent aux 2 graphiques
    fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(6,8)) 

    # premier graphique, objet ax0
    if type_x == "continuous":
        sns.regplot(x=x_var, y=y_var, data=data, lowess=True, ax = ax0)
        ax0.get_lines()[0].set_color('red')
    if type_x == "discrete":
        sns.regplot(x=x_var, y=y_var, data=data, x_estimator=np.mean, ax = ax0)
    if ylim is not None:
        ax0.set_ylim(ylim)
    if xlim is not None:
        ax0.set_xlim(xlim)

    # second graphique : objet ax1
    data[x_var].plot(kind="hist", bins=100, ax=ax1)
    ax1.set_xlabel(xlabel=x_var, fontsize=12)
    ax1.grid(axis="y")
    ax1.set_axisbelow(True)
    if xlim is not None:
        ax1.set_xlim(xlim)
        
    # modification du titre figure : objet 'figure'
    fig.suptitle("Variable {}".format(x_var), fontsize=18)

```

```python
# Source : https://www.kaggle.com/helgejo/titanic/an-interactive-data-science-tutorial
def plot_correlation_map(df, figsize=(13, 12), fontsize=10, savefig=None):
    """
    """
    corr = df.corr()
    fig , ax = plt.subplots(figsize=figsize)
    cmap = sns.diverging_palette( 220 , 10 , as_cmap=True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : fontsize }
    )
    fig.tight_layout()
    fig.savefig(savefig)

regression_graph_quanti_var(data = base_cout, y_var = "claim_amount", x_var = "vh_value", 
                            type_x = "continuous", ylim = [0,2000])
plot_correlation_map(base_cout._get_numeric_data().drop("claim_nb", axis = 1))
```
